
# Titanic Data Analysis

## Objective
The goal of this project is to perform data cleaning and exploratory data analysis (EDA) on the Titanic dataset to uncover insights about passenger demographics, ticket classes, fares, and survival rates.

## Tools & Libraries
- Python
- Pandas
- Matplotlib
- Seaborn



## Dataset Description
The Titanic dataset contains information about passengers on the Titanic ship. The main columns are:

- **PassengerId**: Unique ID for each passenger
- **Survived**: Survival status (0 = No, 1 = Yes)
- **Pclass**: Passenger class (1 = First, 2 = Second, 3 = Third)
- **Name**: Name of the passenger
- **Sex**: Gender of the passenger
- **Age**: Age of the passenger
- **SibSp**: Number of siblings/spouses aboard
- **Parch**: Number of parents/children aboard
- **Ticket**: Ticket number
- **Fare**: Ticket fare
- **Cabin**: Cabin number
- **Embarked**: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)


```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("titanic_cleaned_data.csv")
```

```python
%matplotlib inline
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6,4))
```

```python
sns.countplot(x='Survived', data=df, palette='Set2')
```

```python
plt.title("Survival Count", fontsize=14)
```

```python
plt.xlabel("Survived (0 = No, 1 = Yes)")
```

```python
plt.ylabel("Count")
```

```python
plt.show()
```

```python
### Observation:
- Total passengers: 891
- Survived: Approximately 38%
- Did not survive: Around 62%
- Overall survival rate was quite low.
```

```python
plt.figure(figsize=(8,5))
```

```python
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
```

```python
plt.title("Age Distribution", fontsize=14)
```

```python
plt.xlabel("Age")
```

```python
plt.ylabel("Number of Passengers")
```

```python
plt.show()
```

```python
### Observation:
- Most passengers were between 2035 years old.
- Few children (<10 years) and elderly (>60 years) were on board.
- Age might have played a role in survival chances.
```

```python
plt.figure(figsize=(6,4))
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='pastel')
```

```python
plt.title("Survival by Gender", fontsize=14)
```

```python
plt.xlabel("Gender")
```

```python
plt.ylabel("Count")
```

```python
plt.show()
```

```python
### Observation:
- Female passengers had a much higher survival rate than males.
- Male survival rate was quite low (~18%).
- Likely reason: The "Women and children first" evacuation rule.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='muted')
plt.title("Passenger Class vs Survival", fontsize=14)
plt.xlabel("Class (1 = Best, 3 = Lowest)")
plt.ylabel("Count")
plt.show()
```

```python
### Observation:
- 1st class passengers had the highest survival rate.
- 3rd class passengers had the lowest survival rate.
- Higher class likely meant better cabins and priority evacuation.
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], kde=True, color='orange', bins=40)
plt.title("Fare Distribution", fontsize=14)
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
### Observation:
- Most tickets were priced between 050.
- A few expensive tickets (>400) were found, mainly in 1st class.
- Fare distribution was uneven, showing a clear class difference.
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

```python
### Observation:
- Strong negative correlation (-0.55) between `Pclass` and `Fare`, meaning higher classes had higher ticket prices.
- `Survived` has a moderate positive correlation with `Fare` (~0.26)  passengers who paid higher fares had better survival chances.
- `Survived` has a weak negative correlation with `Pclass` (~-0.34)  higher classes were more likely to survive.
- Age, SibSp, and Parch show very weak correlations with survival, meaning these factors alone did not strongly influence survival chances.
```


## Final Summary & Insights

1. Only about 38% of passengers survived the Titanic disaster.
2. Female passengers had a significantly higher survival rate compared to males.
3. First-class passengers had the highest survival rate, while third-class passengers had the lowest.
4. Most passengers were between 2035 years old.
5. Higher ticket fares were generally associated with better survival chances.
6. There is a strong negative correlation between `Pclass` and `Fare`, showing that higher classes had higher fares.
7. Age, SibSp, and Parch showed weak correlations with survival.
